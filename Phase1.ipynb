Normalizing the images

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
#rescaling the images for normalization
my_generator = ImageDataGenerator(rescale=1./255)
```

```python

my_image_generator = my_generator.flow_from_directory( './DataSet/', target_size=(150, 150), batch_size=6, class_mode='binary')
```

```python
#Prints out a batch of 6 sample images
#Also prints out the images shape
import matplotlib.pyplot as plt
for my_batch in my_image_generator:
   
    images = my_batch[0]
    labels = my_batch[1]
    for i in range(len(labels)):
        print(np.asarray(image).shape)
        plt.imshow(images[i])
        plt.colorbar()
        plt.show()
        print(labels[i])

    break
```

```python

myAugmentedImageGenerator = ImageDataGenerator(
    rotation_range=25,
    width_shift_range=0.35,
    height_shift_range=0.35,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    rescale=1./255,
    fill_mode='nearest')
```

```python
augmentedImageGenerator = myAugmentedImageGenerator.flow_from_directory( './DataSet/', target_size=(255, 255), batch_size=6, class_mode='binary')
```

```python
for my_batch in augmentedImageGenerator:
    images = my_batch[0]
    labels = my_batch[1]
    for i in range(len(labels)):
        plt.imshow(images[i])
        plt.colorbar()
        plt.show()
        print(labels[i])
    break
```
